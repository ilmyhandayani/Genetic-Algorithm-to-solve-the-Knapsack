```python
import random
import numpy as np
import operator

class KnapSack:
  def __init__(self,weights,values,max_weight):
    self.weights = weights
    self.values = values
    self.max_weight = max_weight

  def init(self):
    num_population = 20
    self.population = []
    for i in range(num_population):
      ind = []
      for k in range(len(self.weights)):
        ind.append(random.randint(0,1))
      self.population.append(ind)

  def eval_fit(self,individu):
    weight_sum = np.sum(np.array(self.weights)*np.array(individu))
    value_sum = np.sum(np.array(self.values)*np.array(individu))

    if weight_sum > self.max_weight:
      return False
    else:
      return value_sum

  def adaptation(self):
    new_pop = []
    for id_ind in range(len(self.population)):
      eval = self.eval_fit(self.population[id_ind]) 
      if eval != False:
        new_pop.append((eval,self.population[id_ind]))
    
    # sorting
    new_pop.sort(key=operator.itemgetter(0),reverse=True)
    self.population = [x[1] for x in new_pop]

  def crossover(self):
    parent = random.sample(range(0,len(self.population)),2)
    ind1 = self.population[parent[0]]
    ind2 = self.population[parent[1]]
    # one-point crossover
    th = random.randint(1,len(ind1)-1)
    tmp1 = ind1[th:]
    tmp2 = ind2[th:]
    ind1 = ind1[:th]
    ind2 = ind2[:th]
    ind1.extend(tmp2)
    ind2.extend(tmp1)
    self.population.append(ind1)
    self.population.append(ind2)

  def mutation(self):
    mut = random.randint(0,len(self.population)-1)
    ind = self.population[mut]
    for i in range(len(ind)):
      if random.uniform(0,1) > 0.5:
        if ind[i] == 1:
          ind[i] = 0
        else:
          ind[i] = 1
    self.population[mut] = ind

  def ga(self):
    self.init()
    for i in range(50):
      self.adaptation()
      print("generasi ke ",i,self.population[0],self.eval_fit(self.population[0]))
      self.crossover()
      self.mutation()


weights = [12,7,11,8,9,11,15,22,13,1]
values = [24,13,23,15,16,11,9,5,22,11]
max_weight = 45
# output = [0,1,1,1,1,0,0,0]
KnapSack(weights,values,max_weight).ga()
```
